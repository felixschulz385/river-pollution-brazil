```python
import matplotlib.pyplot as plt
import seaborn as sns

plot_data = results.loc[
    (results["land_cover_subclass"] == "c40")
    & (results["distance_step_name"] == "50_100km")
    & results["term"].str.startswith("lc_", na=False)
    & ~results["term"].str.startswith("streamflow_", na=False)
    & results["pollutant"].isin(["ph", "biochemical_oxygen_demand", "nitrites"])
].copy()

plot_data["coef"] = plot_data["term"].str.replace("_c40_shr__log_0p01", "", regex=False)
plot_data["coef"] = pd.Categorical(
    plot_data["coef"],
    categories=["lc_0_10km", "lc_10_50km", "lc_50_100km"],
    ordered=True,
)
plot_data = plot_data.sort_values(["pollutant", "coef"])

pollutants = plot_data["pollutant"].drop_duplicates().tolist()
ncols = 3
nrows = (len(pollutants) + ncols - 1) // ncols

fig, axes = plt.subplots(
    nrows=nrows,
    ncols=ncols,
    figsize=(4 * ncols, 2.8 * nrows),
    constrained_layout=True,
)
axes = axes.flatten() if hasattr(axes, "flatten") else [axes]
sns.set_theme(style="whitegrid")

for ax, pollutant in zip(axes, pollutants):
    subset = plot_data.loc[plot_data["pollutant"] == pollutant].sort_values("coef")
    ax.errorbar(
        subset["Estimate"],
        subset["coef"].astype(str),
        xerr=[
            subset["Estimate"] - subset["2.5%"],
            subset["97.5%"] - subset["Estimate"],
        ],
        fmt="o",
        color="#0072B2",   # Okabe-Ito blue (color-blind safe)
        ecolor="#56B4E9",  # lighter blue for CI bars
        elinewidth=2,
        capsize=3,
        markersize=6,
    )
    ax.axvline(0, color="black", linestyle="--", linewidth=1)
    ax.set_title(pollutant)
    ax.set_xlabel("Coefficient estimate")
    ax.set_ylabel("Coefficient")

for ax in axes[len(pollutants):]:
    ax.set_visible(False)

# fig.suptitle("Land-cover coefficients for Urbanization at 50_100km", fontsize=16)
fig.savefig("../../output/analysis/sensor_data/land_cover_coefficients_urbanization_50_100km.png", dpi=300)
```
